Following the examples provided in the lecture, create a 3D scatter plot with randomly generated points of variable size and colors. 

Then animate the rotation of 360 degrees around the z-axis. 


```python
%matplotlib notebook
import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
from time import sleep
from mpl_toolkits.mplot3d import Axes3D


rng = np.random.RandomState(0)

# Create figure for plotting
fig = plt.figure(figsize=(4,4))
#plt.ion()

ax = Axes3D(fig)
fig.show()
fig.canvas.draw()

# randomly generate data
x = rng.randn(100)
y = rng.randn(100)
z = 1+2*rng.randn(100)
colors = rng.rand(100)
sizes = 100 * rng.rand(100)

for i in range(120):
    ax.clear()
    ax.scatter(x, y, z, c=colors, s=sizes, alpha=0.5, cmap='viridis')
    ax.view_init(elev=20., azim=-i*3)
    
    # Format plot
    plt.title('Animated 3D plot')
    fig.canvas.draw()  
    #sleep(0.5)
```

```python
# lib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

# Def for creating points
def randrange(n, vmin, vmax):
    l = (vmax - vmin)*np.random.rand(n) + vmin
    return l

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# For each set of style and range settings, plot n random points in the box
n = 100

# Defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for m, zlow, zhigh in [('o', -50, -25), ('^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, marker=m)

# label of axis
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
